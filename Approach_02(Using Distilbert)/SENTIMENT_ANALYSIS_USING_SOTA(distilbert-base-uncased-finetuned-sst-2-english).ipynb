### Application on Raw_data

```python
#!pip install -q transformers
from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis")
#data = ["I love you", "I hate you"] # sample sentiment check
#sentiment_pipeline(data)
```

```python
title_reviews=df['title_reveiws'].tolist()
```

```python
sentiment=pipeline('sentiment-analysis')
data=title_reviews
x=sentiment(data)
x
```

```python
review_list=[]
for i in range(len(x)):
    if x[i]['label']=='NEGATIVE':
        review_list.append(0)
    elif x[i]['label']=='POSITIVE':
        review_list.append(1)
        
print(review_list)
```

```python
df['sentiment_title_review'] = review_list
```

```python
df.head().columns
```

```python
def Sentiment_title_review_tag(x):
    if x==0:
        return 'NEGATIVE'
    elif x==1:
        return 'POSITIVE'
```

```python
df
```

```python
df['Sentiment_title_review_tag']=df['sentiment_title_review'].apply(Sentiment_title_review_tag)
df
```
