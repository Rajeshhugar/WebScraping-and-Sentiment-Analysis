```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('processed_clean_data.csv')
df
```

```python
from transformers import AutoTokenizer,AutoModelForSequenceClassification

from scipy.special import softmax
```

```python
roberta ='cardiffnlp/twitter-roberta-base-sentiment'
```

```python
model = AutoModelForSequenceClassification.from_pretrained(roberta)
```

```python
tokenizer = AutoTokenizer.from_pretrained(roberta)
```

```python
labels=['Negative','Neutral','Positive']
```

```python
title_review=df['title_reveiws'].tolist()
```

```python

#Sentiment  Analysis

encode_review = tokenizer(title_review,return_tensors='pt',padding=True)
```

```python
encode_review
```

```python
output = model(encode_review['input_ids'],encode_review['attention_mask'])
output
```

```python
output[0]
```

```python
scores=[]
for i in range(len(output[0])):
    scores.append(output[0][i].detach().numpy())
    
scores
```

```python
scores[0][0]
```

```python
scores[0]
```

```python
scores_scaler=[]
for i in range(len(scores)):
    x=softmax(scores[i])
    scores_scaler.append(x)
```

```python
scores_scaler[0][0]
```

```python
dict_list={}
for j in range(len(scores_scaler)):
    for i in range(len(labels)):
        l = labels[i]
        s = scores_scaler[0][i]
        print(l,s)
        
    print('**'*25)
```

```python
negative=[]
neutral=[]
positive=[]
for i in range(len(scores_scaler)):
    negative.append(scores_scaler[i][0])
    neutral.append(scores_scaler[i][1])
    positive.append(scores_scaler[i][2])
```

```python
df['negative'] = negative
df['neutral'] = neutral
df['positive'] = positive
df.head()
```

```python
df.columns
```

```python
max_cols = df[['negative','neutral','positive']].idxmax(axis=1)
```

```python
df['Sentiment_having_max_score'] = max_cols
```

```python
df
```

```python
df.to_csv('SENTIMENT_ANALYSIS_USING_twitter-roberta-base-sentiment_on_raw_text.csv',index=False)
```
